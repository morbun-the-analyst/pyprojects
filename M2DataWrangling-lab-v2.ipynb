### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
display(df['Country'].unique())
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
#example dont run check first
import pandas as pd

# Example DataFrame
data = {
    'Country': ['USA', 'US', 'Canada', 'UK', 'USA', 'UK', 'USA'],
    'EdLevel': ['Bachelors', 'Bachelor', 'Masters', 'PhD', 'Bachelors', 'MSc', 'Bachelor']
}
df = pd.DataFrame(data)

# Define a mapping for standardizing Country
country_mapping = {
    'US': 'USA',
    'UK': 'United Kingdom',
    'Canada': 'Canada'
}

# Define a mapping for standardizing EdLevel
edlevel_mapping = {
    'Bachelor': 'Bachelors',
    'Bachelors': 'Bachelors',
    'MSc': 'Masters',
    'Masters': 'Masters',
    'PhD': 'PhD'
}

# Standardize 'Country' and 'EdLevel' columns using replace()
df['Country'] = df['Country'].replace(country_mapping)
df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

# Show the standardized DataFrame
print(df)
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df_encoded = pd.get_dummies(df, columns=['Employment'], prefix='Employment')
display(df_encoded)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_values = df.isnull().sum()

# Sort the columns by the number of missing values in descending order
missing_values_sorted = missing_values.sort_values(ascending=False)

# Show the columns with the highest missing values
print(missing_values_sorted)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
most_frequent_value = df['RemoteWork'].mode()[0]
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent_value)
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
## Write your code here
df['YearsCodePro'].unique()
def categorize_experience(years):
    if pd.isna(years):  # Handle NaN
        return 'Unknown'
    elif years == 'Less than 1 year':
        return 'Beginner'
    elif years == 'More than 50 years':
        return 'Expert'
    else:
        years = int(years)  # Convert to integer
        if years <= 2:
            return 'Beginner'
        elif years <= 5:
            return 'Junior'
        elif years <= 10:
            return 'Intermediate'
        elif years <= 20:
            return 'Advanced'
        else:
            return 'Expert'

# Apply the categorization function
df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)

# Display the DataFrame
display()
```
